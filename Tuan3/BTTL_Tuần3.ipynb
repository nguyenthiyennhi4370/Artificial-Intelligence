```python
#Bài 1
import math

# bàn cờ 3x3
board = [
    [' ', ' ', ' '],
    [' ', ' ', ' '],
    [' ', ' ', ' ']
]

# in bàn cờ
def print_board(board):
    for row in board:
        print("|".join(row))
        print("-" * 5)

# kiểm tra thắng
def check_winner(board):
    # hàng
    for row in board:
        if row[0] == row[1] == row[2] and row[0] != ' ':
            return row[0]

    # cột
    for col in range(3):
        if board[0][col] == board[1][col] == board[2][col] and board[0][col] != ' ':
            return board[0][col]

    # đường chéo
    if board[0][0] == board[1][1] == board[2][2] and board[0][0] != ' ':
        return board[0][0]

    if board[0][2] == board[1][1] == board[2][0] and board[0][2] != ' ':
        return board[0][2]

    return None

# kiểm tra còn nước đi không
def is_moves_left(board):
    for row in board:
        if ' ' in row:
            return True
    return False


# thuật toán minimax
def minimax(board, is_max):
    winner = check_winner(board)

    if winner == 'X':
        return 1
    elif winner == 'O':
        return -1
    elif not is_moves_left(board):
        return 0

    if is_max:
        best = -math.inf

        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = 'X'
                    best = max(best, minimax(board, False))
                    board[i][j] = ' '

        return best

    else:
        best = math.inf

        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = 'O'
                    best = min(best, minimax(board, True))
                    board[i][j] = ' '

        return best


# tìm nước đi tốt nhất cho AI
def best_move(board):
    best_val = -math.inf
    move = (-1, -1)

    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                board[i][j] = 'X'
                move_val = minimax(board, False)
                board[i][j] = ' '

                if move_val > best_val:
                    move = (i, j)
                    best_val = move_val

    return move


# test
board[0][0] = 'O'
board[1][1] = 'X'
board[0][1] = 'O'

print("Bàn cờ hiện tại:")
print_board(board)

move = best_move(board)
print("AI nên đánh tại:", move)
```

```python
#Bài 2
import math

def alphabeta(board, depth, alpha, beta, is_max):
    winner = check_winner(board)

    if winner == 'X':
        return 1
    elif winner == 'O':
        return -1
    elif not is_moves_left(board):
        return 0

    if is_max:
        best = -math.inf

        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = 'X'

                    value = alphabeta(board, depth+1, alpha, beta, False)

                    board[i][j] = ' '
                    best = max(best, value)
                    alpha = max(alpha, best)

                    if beta <= alpha:
                        break

        return best

    else:
        best = math.inf

        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = 'O'

                    value = alphabeta(board, depth+1, alpha, beta, True)

                    board[i][j] = ' '

                    best = min(best, value)
                    beta = min(beta, best)

                    if beta <= alpha:
                        break

        return best


def best_move_alpha_beta(board):
    best_val = -math.inf
    move = (-1, -1)

    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                board[i][j] = 'X'

                move_val = alphabeta(board, 0, -math.inf, math.inf, False)

                board[i][j] = ' '

                if move_val > best_val:
                    best_val = move_val
                    move = (i, j)

    return move


print("Nước đi tốt nhất với Alpha-Beta:")
print(best_move_alpha_beta(board))
```
