```python
#Bài tập về nhà
#Bài 1
def tron_min(a, b):
    c = []
    n = max(len(a), len(b))

    for i in range(n):
        if i < len(a) and i < len(b):
            c.append(min(a[i], b[i]))
        elif i < len(a):
            c.append(a[i])
        else:
            c.append(b[i])

    return c

# Test
a = [5, 10, 1, 4]
b = [2, 7, 3, 7, 9, 8]

print("Kết quả:", tron_min(a, b))
```

```python
#Bài 2
import random

# Nhập kích thước
n = int(input("Nhập n (ma trận vuông): "))

# Tạo ma trận
a = [[random.randint(-10, 10) for j in range(n)] for i in range(n)]

print("Ma trận:")
for row in a:
    print(row)
#a. Tổng tam giác trên của đường chéo phụ
tong = 0
for i in range(n):
    for j in range(n):
        if i + j <= n - 1:
            tong += a[i][j]

print("Tổng:", tong)
#b. Đổi số âm → trị tuyệt đối
for i in range(n):
    for j in range(n):
        if a[i][j] < 0:
            a[i][j] = abs(a[i][j])

print("Sau khi đổi âm:")
for row in a:
    print(row)
#c. Thay số chẵn bằng x
x = int(input("Nhập x: "))

for i in range(n):
    for j in range(n):
        if a[i][j] % 2 == 0:
            a[i][j] = x

print("Sau khi thay:")
for row in a:
    print(row)
#d. Kiểm tra toàn chẵn
toan_chan = True

for i in range(n):
    for j in range(n):
        if a[i][j] % 2 != 0:
            toan_chan = False

print("Toàn chẵn:", toan_chan)
#e. Kiểm tra ma trận đối xứng
doi_xung = True

for i in range(n):
    for j in range(n):
        if a[i][j] != a[j][i]:
            doi_xung = False

print("Đối xứng:", doi_xung)
#f. Kiểm tra đường chéo chính tăng dần
tang_dan = True

for i in range(n - 1):
    if a[i][i] >= a[i+1][i+1]:
        tang_dan = False

print("Đường chéo chính tăng dần:", tang_dan)
#g. Xuất tam giác dưới của đường chéo phụ
print("Tam giác dưới đường chéo phụ:")

for i in range(n):
    for j in range(n):
        if i + j >= n - 1:
            print(a[i][j], end=" ")
    print()
#h. Kiểm tra đường chéo phụ giảm dần
giam_dan = True

for i in range(n - 1):
    if a[i][n-1-i] <= a[i+1][n-2-i]:
        giam_dan = False

print("Đường chéo phụ giảm dần:", giam_dan)

```
