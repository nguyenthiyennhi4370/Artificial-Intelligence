```python
#Bai 14
import random

m = int(input("Nhập số dòng: "))
n = int(input("Nhập số cột: "))

# a. Tạo mảng
a = [[random.randint(1, 10) for j in range(n)] for i in range(m)]

# In mảng
for row in a:
    print(row)

# b. Dòng k
k = int(input("Nhập dòng k: "))
print("Dòng k:", a[k])

# c. Cột k
print("Cột k:", [a[i][k] for i in range(m)])

# d. Dòng có tổng lớn nhất
tong_max = max(sum(row) for row in a)
print("Tổng lớn nhất:", tong_max)

# e. Cột có tích nhỏ nhất
tich_min = float('inf')
for j in range(n):
    tich = 1
    for i in range(m):
        tich *= a[i][j]
    tich_min = min(tich_min, tich)
print("Tích nhỏ nhất:", tich_min)

# f. Dòng chẵn, cột lẻ
print("Phần tử (dòng chẵn, cột lẻ):")
for i in range(0, m, 2):
    for j in range(1, n, 2):
        print(a[i][j], end=" ")
print()

# g. TB cộng số chẵn dòng lẻ
tong = dem = 0
for i in range(1, m, 2):
    for j in range(n):
        if a[i][j] % 2 == 0:
            tong += a[i][j]
            dem += 1
print("TB:", tong/dem if dem != 0 else 0)

# h. TB biên
tong = dem = 0
for i in range(m):
    for j in range(n):
        if i == 0 or i == m-1 or j == 0 or j == n-1:
            tong += a[i][j]
            dem += 1
print("TB biên:", tong/dem)

# i. TB tích không biên
tich = 1
dem = 0
for i in range(1, m-1):
    for j in range(1, n-1):
        tich *= a[i][j]
        dem += 1
print("TB tích:", tich/dem if dem != 0 else 0)
```

```python
import datetime

#Bai 15
class SinhVien:
    def __init__(self, ma, ten, namsinh, diem):
        self.ma = ma
        self.ten = ten
        self.namsinh = namsinh
        self.diem = diem

# b. Tạo danh sách
ds = [
    SinhVien("02DH0001", "Phan Lan", 2005, 6.5),
    SinhVien("01CD0002", "Nguyen An", 2004, 4.0),
    SinhVien("03DH0003", "Tran Lan", 2003, 7.0)
]

# c. Đủ điều kiện
print("SV đủ điều kiện:", sum(1 for sv in ds if sv.diem >= 5))

# d. Đủ 20 tuổi
current_year = datetime.datetime.now().year
print(f"SV >= 20 tuổi (năm hiện tại: {current_year}):")
for sv in ds:
    if current_year - sv.namsinh >= 20:
        print(sv.ten)

# e. Hệ DH
print("SV hệ DH:", sum(1 for sv in ds if sv.ma[2:4] == "DH"))

# f. Tên Lan
print("Tên Lan:", sum(1 for sv in ds if "Lan" in sv.ten))

# g. Họ Phan
print("Họ Phan:", sum(1 for sv in ds if sv.ten.startswith("Phan")))
```
